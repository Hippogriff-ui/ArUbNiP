```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
feature_score_df = pd.read_csv(r"C:\Users\Li Xiaokang\Desktop\null_importances_result\score_df\human_null_scores_win31.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 9))

# 
feature_list = feature_score_df['feature'].tolist()[:50]
y_pos = np.arange(len(feature_list))
scores = feature_score_df['importance'].tolist()[:50]

# 
colors = []
for feature in feature_list:
    if 'PWM' in feature:
        colors.append('#FEB705')
    elif 'CKSAAP' in feature or 'EAAC' in feature or 'Binary' in feature:
        colors.append('#FA8600')
    elif 'AAindex' in feature:
        colors.qppend('#FDB96A')
    else:
        colors.append('#FEE89A')

# 
ax.barh(y_pos, scores, align='center', color=colors, alpha=0.8, edgecolor='#636363', linewidth=1.5)

# 
ax.set_yticks(y_pos)
ax.set_yticklabels(feature_list, fontsize=8)
ax.set_xlabel('final_score', fontsize=12)
ax.set_ylabel('Feature', fontsize=12)

# 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 
ax.invert_yaxis()

# 
plt.tight_layout()

# EPS
plt.savefig('Fig 4-human_actual_score.svg', format='svg', dpi=1200)

# 
plt.show()
```
