```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
actual_imp_df = pd.read_csv(r"C:\Users\Li Xiaokang\Desktop\null_importance_result\train31_RF_actual_nonzero_imp_df.csv")
```

```python
null_imp_df = pd.read_csv(r"C:\Users\Li Xiaokang\Desktop\null_importance_result\train31_RF_null_imp_df.csv")
```

```python
def format_func(x, _):
    return "{:.3f}".format(x*1000)
```

```python
def display_distributions(actual_imp_df_, null_imp_df_, feature_,fig_name):
    plt.figure(figsize=(6, 4))

    a = plt.hist(null_imp_df_.loc[null_imp_df_['feature'] == feature_, 'importance'].values, label='null_scores distribution',color='#4B74B2')
    plt.vlines(x=actual_imp_df_.loc[actual_imp_df_['feature'] == feature_, 'importance'].mean(), 
               ymin=0, ymax=np.max(a[0]), color='#DB3124',linewidth=5, label='actual_score')
    
    plt.xlabel('score (10$^{-3}$)')
    plt.ylabel('Frequency')  # y-axis label
    legend = plt.legend(loc='best')
    legend.get_frame().set_alpha(0.3)
    legend.get_texts()[0].set_fontsize(10)
    legend.get_texts()[1].set_fontsize(10)
    
    plt.gca().xaxis.set_major_formatter(FuncFormatter(format_func))

    plt.savefig(fig_name, format='pdf', dpi=1200)
```

```python
display_distributions(actual_imp_df_=actual_imp_df, null_imp_df_=null_imp_df, feature_='AA531_7003',fig_name='Fig 3-AA531_7003.pdf')
```
