```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
actual_imp_df = pd.read_csv(r"D:\study\paper\ubiquitination\null_importances_result\score_df\Arab_actual_nonzero_score_win31.csv")
```

```python
null_imp_df = pd.read_csv(r"D:\study\paper\ubiquitination\null_importances_result\score_df\Arab_null_scores_win31.csv")
```

```python
def format_func(x, _):
    return "{:.3f}".format(x*1000)
```

```python
def display_distributions(actual_imp_df_, null_imp_df_, feature_,fig_name):
    plt.figure(figsize=(6, 5))
    
    # 设置默认字体样式和大小
    plt.rcParams['font.sans-serif'] = 'Arial'
    plt.rcParams['font.size'] = 12
    a = plt.hist(null_imp_df_.loc[null_imp_df_['feature'] == feature_, 'importance'].values, label='null scores distribution',color='#898ca9')
    plt.vlines(x=actual_imp_df_.loc[actual_imp_df_['feature'] == feature_, 'importance'].mean(), 
               ymin=0, ymax=np.max(a[0]), color='#de425b',linewidth=5, label='actual score')
    
    plt.xlabel('score (10$^{-3}$)')
    plt.ylabel('Frequency')  # y-axis label
    legend = plt.legend(loc='best')
    legend.get_frame().set_alpha(0.3)
    legend.get_texts()[0]
    legend.get_texts()[1]
    
    plt.gca().xaxis.set_major_formatter(FuncFormatter(format_func))
    plt.savefig(fig_name, format='pdf', dpi=1200)
```

```python
display_distributions(actual_imp_df_=actual_imp_df, null_imp_df_=null_imp_df, feature_='EAAC_195',fig_name='EAAC_195.pdf')
```
